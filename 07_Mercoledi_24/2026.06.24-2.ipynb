```python
lista = [1,2,3,5,6,7]

def funz(a):
    return a+5

lista2 = list(map(funz, lista))
#map, non returna true o false, ma applica quella funzione alla lista passata


print(lista2)
```

```python
lista = [1,2,3,5,6,7]

def funz(a):
    return a%2==0

lista2 = []
for el in lista:
    if funz(el):
        lista2.append(el)

print(lista2)
```

```python
lista = [1,2,3,5,6,7]

def funz(a):
    return a%2==0

lista2 = list(filter(funz,lista))   #lista2 prende lista a cui è stata applicata la funz, eleme
#applcica funz a tutti elementi lista, funziona con true e false, se vero inserisce

print(lista2)
```

```python
lista = [1,2,3,5,6,7]


lista2 = list(map(lambda num:num+5, lista)) #applico funzione alla lista elemento per elemento
#map invece, non returna true o false, ma applica quella funzione alla lista passata


print(lista2)
```
