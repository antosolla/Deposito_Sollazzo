```python
lista = ["tommaso","teresa","maria"]

lista1 = [5,2,10]

for nome, ordini in zip(lista,lista1):
    print(f"nome cliente: {nome}, n.ordini: {ordini}")
```

```python
lista = ["tommaso","teresa","maria"]

lista1 = [5,2,10]

diz = {}

for nome, ordini in zip(lista,lista1):
    diz.setdefault(nome,ordini)

print(diz)
```
