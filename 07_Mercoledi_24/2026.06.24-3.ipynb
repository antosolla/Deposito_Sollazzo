```python

while True:
    numero = input("inserisci un numero: ")
    try:
        print(5/int(numero))
        break
    except ValueError as e:
        print("numero non valido, inserire un numero corretto!",e)
    except ZeroDivisionError as e:
        print("non si può dividere per 0",e)
    except:
        print("errore non riconosciuto")

print("proseguo programma")
```

```python
with open("file.csv","r") as file:
    contenuto = file.read()
print(contenuto,"\n")
righe = contenuto.split("\n")

matrice = [riga.split(",") for riga in righe]

for riga in matrice:
    print(riga)
```

```python
#m = len(matrice)
#n =len(righe[0])

for riga in range(len(matrice)):
    #for col in range(len(matrice[riga])):
    if matrice[riga][1] == "rossi":
        matrice[riga][1] = "sollazzo"
        break

for riga in matrice:
    print(riga)
```

```python
listaRigheMod= []
for riga in matrice:
    listaRigheMod.append(",".join(riga))

print(listaRigheMod)
```

```python
stringa = "\n".join(listaRigheMod)

print(stringa)
```

```python
with open("file.csv","w") as file:
    file.write(stringa)
```
