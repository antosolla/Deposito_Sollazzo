
create un programma python che prende in input tre numeri e ci dice:
- se almeno 2 di questi numeri sono uguali
- altrimenti il numero più grande dei tre


```python
a = input("Inserisci il primo numero\n")
b = input("Inserisci il secondo numero\n")
c = input("Inserisci il terzo numero\n")
max_numero = a

if a == b:
    print("a e b sono uguali")
elif a == c:
    print("a e c sono uguali")
elif b == c:
    print("b e c sono uguali")
else:
    print("NON Ci sono almeno due numeri uguali")
    if b > max_numero:
        max_numero = b
    if c > max_numero:
        max_numero = c
    
    print(f"Il numero piu grande è: {max_numero}" )
```

Il programma fornisce suggerimenti (troppo alto, troppo basso), termina quando l'utente indovina correttamente, quando i tentativi finiscono o se
scrive «mi arrendo».

```python
print("giocatore 1, inserisci il numero (compreso tra 0 e 100)\n")
while True:
    numero = input()
    if numero > 0 and numero <= 100:
        break
    else:
        print("Input non valido")
    

count = 1
flag = False

while count <=5:
    guess = input("Prova a indovinare:")
    if guess == "exit":
        break
    elif guess == numero:
        print("Hai indovinato!")
        flag = True
        break
    elif guess > numero:
        print("Il numero da indovinare è piu basso")
    else:
        print("Il numero da indovinare è piu alto")
        
    count = count + 1
if not flag:
    print("Peccato. Non hai indovinato") 
```
